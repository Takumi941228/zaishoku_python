```python
# Numpyモジュールをnpという名前でimport
import numpy as np

# グラフ描画ライブラリ matplotlibのpyplotを plt という名前でimport
import matplotlib.pyplot as plt

#Jupyter Notebookの場合は、描画グラフをブラウザ内に表示するために inline指定
%matplotlib inline
```

```python
x = [0, 1, 2, 3, 4, 5]
y = [0, 2, 4, 6, 8, 10]
```

```python
plt.plot(x, y)
```

```python
x_data =  [0, 1, 2, 3, 4, 5]
y_data = [2, 4, 9, 5, 1, 8]
```

```python
#折れ線グラフ
plt.plot(x_data, y_data, color="r")
```

```python
x_data = np.array([1, 2, 3, 4, 5])
y_data = np.array([12, 14, 15, 9, 8])
```

```python
# 棒グラフ
plt.bar(x_data, y_data)
```

```python
#-5から5まで0.1の増え幅で配列を作る
x = np.arange(0, 2*3.14, 0.1)
```

```python
# sin(x)をyのlist格納して生成
y = np.sin(x)
```

```python
# sinカーブの描画
plt.plot(x, y)
```

補足プログラム

```python
#グラフタイトルの追加
plt.title('sine wave')

#x軸とy軸ラベルの追加
plt.xlabel('X')
plt.ylabel('Y')

# 赤色に変更（r=red）
plt.plot(x, y, color='r')
```
