```python
import numpy as np

# グラフ描画ライブラリ matplotlibのpyplotを plt という名前でimport
import matplotlib.pyplot as plt

#Jupyter Notebookの場合は、描画グラフをブラウザ内に表示するために inline指定
%matplotlib inline
```

```python
#-5から5まで0.1の増え幅で配列を作る
x = np.arange(-5, 5, 0.1)

x
```

```python
y = np.sin(x)
```

```python
plt.plot(x, y)
```

```python
#グラフタイトルの追加
plt.title('sine wave')

#x軸とy軸ラベルの追加
plt.xlabel('X')
plt.ylabel('Y')

plt.plot(x, y)
```

```python
#折れ線グラフ
x_data = np.array([1, 2, 3, 4, 5])
y_data = np.array([2, 5, 3, 7, 4])
```

```python
plt.plot(x_data, y_data)
```
