演習プログラム１

```python
# グラフ描画ライブラリ matplotlibのpyplotを plt という名前でimport
import matplotlib.pyplot as plt

#Jupyter Notebookの場合は、描画グラフをブラウザ内に表示するために inline指定
%matplotlib inline
```

```python
# x軸のリストとy軸のリストに数値を格納
x = [0, 1, 2, 3, 4, 5]
y = [0, 2, 4, 6, 8, 10]
```

```python
# グラフ描画
plt.plot(x, y)
```

```python
# x軸のリストとy軸のリストに数値を格納
x_data1 =  [0, 1, 2, 3, 4, 5]
y_data1 = [2, 4, 9, 5, 1, 8]
```

```python
#折れ線グラフ
plt.plot(x_data1, y_data1)
```

```python
# Numpyをnpという名前でimport
import numpy as np
```

```python
#0から6.2まで0.1の増え幅でnumpy配列を作る
x_data2 = np.arange(0, 2*3.14, 0.1)

x_data2
```

```python
# sin(x)をyのlist格納して生成
y_data2 = np.sin(x_data2)

y_data2
```

```python
# sinカーブの描画
plt.plot(x_data2, y_data2)
```

演習プログラム２

```python
#グラフタイトルの追加
plt.title('sine wave')

#x軸とy軸ラベルの追加
plt.xlabel('X')
plt.ylabel('Y')

# 赤色に変更（r=red）
plt.plot(x_data2, y_data2, color='r')
```

```python
# x軸のarrayとy軸のarrayに数値を格納
x_data3 = np.array([1, 2, 3, 4, 5])
y_data3 = np.array([12, 14, 15, 9, 8])
```

```python
# 棒グラフ
plt.bar(x_data3, y_data3)
```

補足プログラム

```python
# 空のサブプロットを作成
# fig = Figure:描画領域全体, axes = Axes:一つのプロットを描画する領域
# axesはAxesオブジェクトの2x2の配列
# figsize=(横, 縦)の比率で大きさを指定
fig, axes= plt.subplots(2,2,figsize=(15,10))

# サブプロットにタイトル追加
axes[0,0].set_title("subplot_0")
axes[0,1].set_title("subplot_1")
axes[1,0].set_title("subplot_2")
axes[1,1].set_title("subplot_3")
 
# サブプロットに軸ラベル追加
axes[0,0].set_xlabel("x-label")
axes[0,0].set_ylabel("y-label")
axes[0,1].set_xlabel("x1-label")
axes[0,1].set_ylabel("y1-label")
axes[1,0].set_xlabel("x2-label")
axes[1,0].set_ylabel("y2-label")
axes[1,1].set_xlabel("x3-label")
axes[1,1].set_ylabel("y3-label")

# 上記4つのグラフを描画
axes[0,0].plot(x,y)
axes[0,1].plot(x_data1,y_data1)
axes[1,0].plot(x_data2,y_data2)
axes[1,1].bar(x_data3,y_data3)
```
