```python
# pandasライブラリをpsという名前でのimport
import pandas as pd

# グラフ描画ライブラリ matplotlibのpyplotを plt という名前でimport
import matplotlib.pyplot as plt

#Jupyter Notebookの場合は、描画グラフをブラウザ内に表示するために inline指定
%matplotlib inline
```

```python
# カラムに名前を付けて、csvファイルを読み込む。文字化けしないようにutf8にエンコードしている
# 気象庁より宇都宮、佐野、小山、黒磯の方面の8月14日～9月14日の一か月間の温度をcsv形式でダウンロード
df = pd.read_csv('data_sample.csv', names=("Data", "Utsunomiya", "Sano", "Oyama", "Kuroiso"),encoding='utf8')

df
```

```python
# Pandasのデータフレームを使ってグラフを自動生成
df.plot()
```

補足プログラム

```python
# サブプロットを作成
# fig = Figure:描画領域全体, ax = Axes:一つのプロットを描画する領域
fig, ax = plt.subplots()

#軸ラベルの追加およびタイトルの追加
ax.set_xlabel("Data")
ax.set_ylabel("Temperature[℃]")
ax.set_title("Average Temperature")

#描画サイズの変更
df.plot(ax=ax, figsize=(10,5))
```

```python
# figureの中に2×2のaxesを生成
# figsizeは縦、横の大きさを指定
fig, axes = plt.subplots(2,2,figsize=(15,7))

#各indexとカラムを指定して、それぞれの場所にプロット
df.plot(ax=axes[0,0], x='Data', y=["Utsunomiya"], color="r")
df.plot(ax=axes[0,1], x='Data', y=["Sano"], color="b")
df.plot(ax=axes[1,0], x='Data', y=["Oyama"], color="g")
df.plot(ax=axes[1,1], x='Data', y=["Kuroiso"], color="y")
```
